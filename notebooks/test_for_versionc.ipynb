```python
import requests
import pandas as pd
from io import StringIO
```

```python

# ---- Fill these in ----
node_id = "78220000150161"
start_time = "2023-01-01T00:00:00Z"
end_time = "2023-02-01T00:00:00Z"
fields = "time,wse"

# Hydrocron endpoint (PO.DAAC)
url = (
    "https://soto.podaac.earthdatacloud.nasa.gov/hydrocron/v1/timeseries"
    f"?feature=Node"
    f"&feature_id={node_id}"
    f"&start_time={start_time}"
    f"&end_time={end_time}"
    f"&output=csv"
    f"&fields={fields}"
)
hydrocron_url_template = (
        "https://soto.podaac.earthdatacloud.nasa.gov/hydrocron/v1/timeseries"
        "?feature=Node&feature_id={node_id}"
        "&start_time={start_time}&end_time={end_time}"
        "&output=csv"
        "&fields={fields}"
    )
response = requests.get(url)
response.raise_for_status()

# Read into pandas
df = pd.read_csv(StringIO(response.text))

print("\nColumns returned:")
print(df.columns.tolist())

# Look for likely version indicators
possible_version_cols = [
    col for col in df.columns
    if "version" in col.lower() or "collection" in col.lower()
]

if possible_version_cols:
    print("\nVersion-related fields found:")
    for col in possible_version_cols:
        print(f"{col}: {df[col].iloc[0]}")
else:
    print("\nNo explicit version column found in CSV output.")
    print("Check HTTP headers instead:")

    for key, value in response.headers.items():
        if "version" in key.lower() or "collection" in key.lower():
            print(f"{key}: {value}")
```
